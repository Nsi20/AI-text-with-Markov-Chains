```python
descriptions = data[1]

corpus = " ".join(descriptions.dropna())

print(corpus[:500])
```

```python
import re

corpus = corpus.lower()

corpus = re.sub(r"[^a-zA-Z0-9\s]", "", corpus)

tokens = corpus.split()

print(tokens[:20])
```

# **Building the Markov Chain Model**

```python
import random
from collections import defaultdict

markov_chain = defaultdict(list)

for current_token, next_token in zip(tokens, tokens[1:]):
    markov_chain[current_token].append(next_token)

def generate_text(chain, start_token, length=50):
    current_token = start_token
    text = [current_token]

    for _ in range(length - 1):
        if current_token not in chain or not chain[current_token]:
            break

        next_token = random.choice(chain[current_token])
        text.append(next_token)
        current_token = next_token

    return " ".join(text)


start_word = "motivational"
generated_text = generate_text(markov_chain, start_word)
print(generated_text)
```

# **Upgrading to N-Grams for Better Context**

```python
trigram_chain = defaultdict(list)

for i in range(len(tokens) - 2):
    current_bigram = (tokens[i], tokens[i + 1])
    next_word = tokens[i + 2]
    trigram_chain[current_bigram].append(next_word)

def generate_text_with_trigrams(chain, start_bigram, length=50):
    current_bigram = start_bigram
    text = list(current_bigram)

    for _ in range(length - 2):
        if current_bigram not in chain or not chain[current_bigram]:
            break

        next_word = random.choice(chain[current_bigram])
        text.append(next_word)
        current_bigram = (current_bigram[1], next_word)

    return " ".join(text)


start_bigram = ("motivational", "speaker")
generated_text = generate_text_with_trigrams(trigram_chain, start_bigram)
print(generated_text)
```

# **Fine-Tuning and Next Steps**

```python
def generate_text_with_trigrams(chain, start_bigram, length=50):
    current_bigram = start_bigram
    text = list(current_bigram)

    for _ in range(length - 2):
        if current_bigram in chain and chain[current_bigram]:

            next_word = random.choice(chain[current_bigram])
        else:

            next_word = random.choice(tokens)
        text.append(next_word)
        current_bigram = (current_bigram[1], next_word)

    return " ".join(text)
```

```python
start_bigram = ("motivational", "speaker")
generated_text = generate_text_with_trigrams(trigram_chain, start_bigram, length=100)
print(generated_text)
```

# **Using GPT-2 for text generation:**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

prompt = "motivational speaker and his message on success"
inputs = tokenizer.encode(prompt, return_tensors="pt")

outputs = model.generate(inputs, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
